```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import laspy
import scipy.spatial as spatial
# %matplotlib widget
```

```python
def find_neighbors(pindex, triang):
    return triang.vertex_neighbor_vertices[1][triang.vertex_neighbor_vertices[0][pindex]:triang.vertex_neighbor_vertices[0][pindex+1]]
```

```python
inFile = laspy.file.File("points.las", mode = "r")
```

```python
points = np.array([inFile.x,inFile.y,inFile.z]).T
```

```python
# Options: http://www.qhull.org/html/qh-optq.htm
tri = spatial.Delaunay(points, qhull_options = "Qbb Qc Qz")
```

```python
plt.close("all")
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for pidx in range(0, points.size):
    neighbors = find_neighbors(pidx, tri)
    if neighbors.size > 0:
        break

for p in tri.points[neighbors]:
    ax.plot([p[0],tri.points[pidx,0]],[p[1],tri.points[pidx,1]],[p[2],tri.points[pidx,2]],'k')
    
plt.show()
```
